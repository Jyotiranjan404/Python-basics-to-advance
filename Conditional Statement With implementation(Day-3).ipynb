```python
#Given an integer input perform the following conditional actions:

#If n is odd, print odd
#If n is even and in the inclusive range of 2 to 5, print small
#If n is even and in the inclusive range of 6 to 20, print medium
#If n is even and greater than 20, print big

Num=int(input("Enter a number"))
if Num%2==0:
    if Num in range(2,5):
        print("Small number")
    elif Num in range(6,20):
        print("median number")
    else:
        print("large number")
else:
    print("number is odd")
```

# While Loop

```python
# print first 10 numbers 
i=1
while i<=10:
    print(i)
    i+=1
```

```python
# find average and sum for first 10 numbers 
count = 1
sum1 = 0
avg = 0
while count <= 10:
    sum1 += count
    count += 1
    avg = sum1 // count
print('Sum     = ',sum1)
print('Average = ',avg)
```

```python
#Draw a line using while loop 
i = 1
while i <= 50:
    print('_', end= ' ')
    i += 1
```

```python
# Another way to draw a line
print("_________________________________________________________________________________")
print("---------------------------------------------------------------------------------")
print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
print("#################################################################################")
```

```python
#calculate the sum of m to n  numbers using input
m=int(input("Enter a number"))
n=int(input("Enter another number"))
Sum=0
while m<=n:
    Sum+=m
    m+=1
print(Sum)
```

```python
# countdown 10 to 0 using input
countdown=int(input("Enter a number"))
while countdown>=0:
    print(countdown)
    countdown-=1  
```

```python
#Enter a number, run a while loop and add all numbers
i=int(input("enter a number"))
Sum=0
while i>=0:
    print(i)
    Sum+=i
    i-=1
print("sum of number is=",Sum)
```

# For loop

```python
#print even and odd for range of number

for i in range(1,10):
    if i % 2 == 0:
        print('{} is {}'.format(i,'even'))
    else:
        print('{} is {}'.format(i,'odd'))
```

```python
#Multiplication table using input
n = int(input("Input a number: "))
for i in range(1,11):
    print(n,'x',i,'=',n*i)
```

```python
# break statement: if codition is matched then break the statement and stop
for val in "priyanka":
    if val == "i":
        break
    print(val)
```

```python
#countinue statement:if codition is matched then remove that part and execute further
food=["Briyani","Vada","Poori","Spinach","Egg","Juice"]
for x in food:
    if(x=="Spinach"):
        continue
    print(x)
```
